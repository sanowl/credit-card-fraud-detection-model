Spliting the data into featuers & targets


```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']


```

```python
print(X)
```

```python
print(Y)
```

Spilt the data into training data and testing data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)


```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training 

Logistic regression Model

```python
model = LogisticRegression()
```

```python
# traing the LogisticRegression with the  training data
model.fit(X_train,Y_train)
```

Accuracy socre


```python
# accuracy on traing data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on traing data:',training_data_accuracy)
```

```python
# accury on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy =  accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accury score on test data:',test_data_accuracy)
```
